# Prática 7: Algoritmo Clonal
Na prática a seguir implementamos um algoritmo clonal para encontrar os melhores valores de x e y para encontrar os mínimos da função de teste de otimização de mishra, a bird function.

Inicialmente vamos instanciar a classe criada:

```python
from clonal import ClonalAlgorithym
import numpy as np
import matplotlib.pyplot as plt
```

```python
#inicialização da classe Clonal
ca = ClonalAlgorithym()
w = ca.run(sample_size=50,n_iterations=50,cloning_rate=5,mutation_rate=10)
```

```python
def plot_population(population, index):
    fig = plt.figure(figsize=(10, 7))
    x = np.linspace(-10, 10, 100)
    y = x
    X, Y = np.meshgrid(x, y)
    Z = np.sin(X)*np.exp(np.power(1-np.cos(Y), 2))+np.cos(Y)*np.exp(np.power(1-np.sin(X), 2))+np.power((X-Y), 2)
    plt.figure(2+index)
    plt.contour(X, Y, Z)
    x1 = [x[0] for x in population]
    x2 = [x[1] for x in population]
    best_x1 = x1[0]
    best_x2 = x2[0]
    plt.plot(best_x1, best_x2, marker='o', color='red')
    plt.scatter(x1, x2, color='blue')
    plt.title('População final')

for i in range(len(w)):
    plot_population(w[i], i)

plt.show()
plt.clf()
plt.close()
```

Como pode-se observar, o algoritmo funcionou de forma esperada, onde os pares de valores selecionados convergiram com sucesso para os vários mínimos locais espalhados pela bird function, apresentando o comportamento esperado para um algoritmo imunológico multimodal
